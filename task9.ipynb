```python
#import required libraries
import matplotlib.pylab as plt
import seaborn as sns
import pandas as pd
```

```python

df=pd.read_csv(r"C:\Users\prana\OneDrive\Desktop\tekworks\day9\Practice-Datasets .csv")
df
```

```python
#1. Create a histogram to analyze the distribution of Age
plt.figure(figsize=(5,5))
plt.hist(df['Age'], bins=10, color='yellow',edgecolor='red')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

#2. Plot a pie chart showing the percentage distribution of Gender
plt.figure(figsize=(5,5))
colors = sns.color_palette('pastel')[0:2]
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%',colors=colors)
plt.ylabel('')
plt.title('Gender Distribution')
plt.show()
```

```python

#3 Create a bar chart showing the number of customers in each City
plt.figure(figsize=(5,5))
sns.countplot(data=df,x='City',color='purple',edgecolor='red')
plt.title('Number of Customers in Each City')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

```python

#4. Plot a box plot to analyze Monthly Income distribution
plt.figure(figsize=(5,5))
df.boxplot(column='Monthly_Income')
plt.title('Monthly Income')
plt.show()
```

```python

#5 . Create a histogram of Spending Score and interpret the skewness
plt.figure(figsize=(5,5))
plt.hist(df['Spending_Score'], bins=10,color='green',edgecolor='black')
plt.title('Spending Score Distribution')
plt.show()
```

```python
#6. Plot a line chart to visualize Purchase Amount trend over Purchase Date
plt.figure(figsize=(5,5))
df.groupby('Purchase_Date')['Purchase_Amount'].sum().plot(kind='line')
plt.xlabel('Purchase Date')
plt.ylabel('total Purchase Amount')   
plt.title('Purchase Amount Over Time')
plt.show()
```

```python

#7. Create a bar plot to compare average Monthly Income by Gender
plt.figure(figsize=(5,5))
sns.barplot(x='Gender',color='pink',edgecolor='black', y='Monthly_Income', data=df)
plt.title('Average Monthly Income by Gender')
plt.show()
```

```python

#8. Plot a box plot of Purchase Amount grouped by Product Category
plt.figure(figsize=(5,5))
sns.boxplot(x='Product_Category', y='Purchase_Amount', data=df)
plt.xticks(rotation=20)
plt.show()
```

```python

#9. Create a pie chart to show Product Category distribution
plt.figure(figsize=(5,5))
df['Product_Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Product Category Distribution')
plt.ylabel('')
plt.show()
```

```python

#10. Plot a histogram of Purchase Amount with appropriate bin size
plt.figure(figsize=(5,5))
plt.hist(df['Purchase_Amount'], bins=30,color='brown',edgecolor='yellow')
plt.title('Purchase Amount Distribution')
plt.show()
```

```python

#11. Create a bar chart showing average Spending Score by City
plt.figure(figsize=(5,5))
sns.barplot(x='City', y='Spending_Score',color='grey',edgecolor='black', data=df, errorbar=None)
plt.show()
```

```python


plt.figure(figsize=(5,5))

df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

df['Month'] = df['Purchase_Date'].dt.to_period('M')

monthly_sales = df.groupby('Month')['Purchase_Amount'].sum()

monthly_sales.plot(kind='line')

plt.xlabel('Month')
plt.ylabel('Total Purchase Amount')
plt.title('Total Purchase Amount per Month')

plt.show()
```

```python

#13. Create a box plot comparing Monthly Income across different Cities
plt.figure(figsize=(5,5))
sns.boxplot(x='City', y='Monthly_Income', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python

#14. Plot a histogram to analyze Age distribution by Gender
plt.figure(figsize=(5,5))
sns.histplot(data=df, x='Age', hue='Gender', bins=10, kde=True)
plt.title('Age Distribution by Gender')
plt.show()
```

```python

#15. Create a bar chart showing total Purchase Amount by Product Category
plt.figure(figsize=(5,5))
df.groupby('Product_Category')['Purchase_Amount'].sum().plot(kind='bar')
plt.xticks(rotation=45)
plt.title('Total Purchase Amount by Product Category')
plt.show()
```

```python

#16. Plot a scatter plot between Monthly Income and Spending Score
plt.figure(figsize=(5,5))
sns.scatterplot(x='Monthly_Income', y='Spending_Score', data=df)
plt.title('Income vs Spending Score')
plt.show()
```

```python

#17. Create a line chart to visualize daily Purchase Amount variation
daily = df.groupby(df['Purchase_Date'].dt.date)['Purchase_Amount'].sum()
plt.figure(figsize=(5,5))
daily.plot()
plt.xticks(rotation=45)
plt.title('Daily Purchase Amount Variation')
plt.show()
```

```python

#18. Plot a box plot of Spending Score grouped by Gender
plt.figure(figsize=(5,5))
sns.boxplot(x='Gender', y='Spending_Score', data=df)
plt.title('Spending Score by Gender')
plt.show()
```

```python

#19. Create a bar chart showing customer count by Product Category and Gender
plt.figure(figsize=(5,5))
df.groupby(['Product_Category', 'Gender']).size().unstack(fill_value=0).plot(kind='bar')
plt.xlabel('Product Category')
plt.ylabel('Customer Count')
plt.title('Customer Count by Product Category and Gender')
plt.legend(title='Gender')
plt.show()
```

```python

#20. Analyze the relationship between Age and Purchase Amount using a suitable visualization
plt.figure(figsize=(5,5))
sns.scatterplot(x='Age', y='Purchase_Amount', data=df)
plt.title('Age vs Purchase Amount')
plt.show()
```
